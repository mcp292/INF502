```python
class Movie:
    
    movie_list = []
    def __init__(self,title="",genres=[],year=0,my_review="",actors=[],watch_cnt=0,borrowed=False,borrower_name=""):
        self.title=title
        self.genres=genres
        self.year=year
        self.my_review=my_review
        self.actors=actors
        self.watch_cnt=watch_cnt
        self.borrowed=borrowed
        self.borrower_name=borrower_name
        Movie.movie_list.append(self)

    def watch_movie(self):
        self.watch_cnt +=1

    def write_review(self,review):
        self.my_review=review

    def set_title(self,title):
        self.title=title

    def set_genres(self,genres):
        self.genres=genres

    def set_year(self,year):
        self.year=year

    def set_actors(self,actors):
        self.actors=actors

    def borrow_movie(self,borrower):
        self.borrower_name=borrower
        self.borrowed=True

    def return_movie(self):
        self.borrower_name=""
        self.borrowed=False

    def list_details(self):
        print("Movie: "+self.title+"    Year: "+str(self.year))
        print("Genre: ",end="")
        print(*self.genres, sep=", ")
        print("List of Actors:")
        for actor in self.actors:
            print("    "+actor.name)

    def print_borrowing_situation(self):
        print("Borrower name: "+self.borrower_name)
        print("Borrowed Flag: "+str(self.borrowed))
```

```python
class Actor:
    def  __init__(self,name="",date_of_birth="",nationality=""):
        self.name=name
        self.date_of_birth=date_of_birth
        self.nationality=nationality
    def __str__(self):
        return ("Name: "+self.name+"\n"+"Date of Birth: "+self.date_of_birth+"\n"+"Nationality: "+self.nationality) 

    def set_name(self,name):
        self.name=name

    def set_date_of_birth(self,date_of_birth):
        self.date_of_birth=date_of_birth

    def set_nationality(self,nationality):
        self.nationality=nationality

    def print_movies(self):
        print("Movies that "+self.name+" participated")
        for movie in Movie.movie_list:
            if self in movie.actors:
                print(movie.title)
```
